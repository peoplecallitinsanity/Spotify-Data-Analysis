```python
#Importing libraries and dataset
import pandas as pd
import numpy as np
df = pd.read_csv('Sigmoid_Spotify_Data_2.csv')
```

# Artists

```python
#Selecting song listened to more than 10s and reseting index
df2 = df[df['msPlayed'] > 10000]
df2.reset_index(drop=True,inplace=True)
df2
```

```python
#Removing the duplicates of artists and track from the same listener
df3 = df2[~pd.DataFrame(np.sort(df2[['sigmaritan','artistName', "trackName"]], axis=1)).duplicated(keep='first')]
```

```python
#Grouping by sigmaritans and getting the most common artist names
df3 = df3.groupby(['sigmaritan'])['artistName'].value_counts()
```

```python
#Creating a list with names
sigmaritans = df['sigmaritan'].unique()
#Saving in a txt file the results
with open("5_most_listened_artists_sigmaritan.txt", "w",encoding='utf8') as f:
    for name in sigmaritans:
        print(f'\n----------{name}---------\n', file=f)
        print(df3[name][:6], file=f)
```

# Songs

```python
#Adding a new column to make the work easier
df2['song'] = df2['artistName'] +' '+ df2['trackName']
```

```python
del df3 
#Grouping by sigmaritans and getting the most common songs
df3 = df2.groupby(['sigmaritan'])['song'].value_counts()
```

```python
#Saving the results in a txt file
with open("5_most_listened_songs_sigmaritan.txt", "w",encoding='utf8') as f:
    for name in sigmaritans:
        print(f'\n----------{name}---------\n', file=f)
        print(df3[name][:5], file=f)
```

# Longest songs

```python
#Importing dataset
duration = pd.read_csv('duration_df.csv')
duration
```

```python
#Selecting only the tracks that were listened to almost entirely (>=0.9 of duration)
duration = duration.loc[duration['msPlayed'] / duration['duration'] >= 0.9]
```

```python
#function to transform miliseconds in minutes
def ms_to_min(x):
    return x/60000
```

```python
#applying the function
duration['minutes'] = duration['duration'].apply(ms_to_min)
```

```python
#Reseting index
duration.reset_index(inplace = True, drop = True)
```

```python
duration
```

```python
#Grouping by sigmaritan and sorting duration column descending
df3 = duration.groupby(['sigmaritan']).apply(lambda x: x.sort_values('duration',ascending=False))
```

```python
#Reseting index
df3 = df3.reset_index(drop=True)
```

```python
#Removing duplicates of songs by same listener and keeping the first one(the longest listened to)
df3 = df3[~pd.DataFrame(np.sort(df3[['sigmaritan','trackName', 'artistName']], axis=1)).duplicated(keep='first')]
```

```python
#Grouping by sigmaritan and getting the first 5 songs
df3 = df3.groupby(['sigmaritan']).head(5)
```

```python
df3
```

```python
#Saving results
df3.to_csv('longest_5_tracks.csv', index = True)
```
